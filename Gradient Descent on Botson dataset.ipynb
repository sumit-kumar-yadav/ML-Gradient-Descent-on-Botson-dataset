```python
# Import the reqd libraries
import numpy as np
import pandas as pd
```

```python
# Load the botson data fron the given csv file
df = pd.read_csv("train.csv")
print(df.shape)
df.head()
```

```python
# Check if there is a null value
print(df.isnull().sum())
```

```python
# Delete the output column --- >> coulmn Y and store it to Y
print(df.columns, df.shape)
Y = df[' Y']  # Storing the output
del df[' Y']
print(df.columns, df.shape)
```

```python
# Add the last column having value = 1 so that we can apply Gradient Descent
df['c'] = np.ones(df.shape[0])
```

```python
print(df.shape)
df.head()
```

```python
# Convert df to np array
data = np.array(df)
print(data)
print(data.shape)
```

```python
def step_gradient(points, learning_rate, m):
    # Initialize the value
    cols = points.shape[1]
    m_slope = np.zeros(cols)
    M = len(points)
    
    # Calculate the values
    for i in range(M):
        x = points[i]
        y = Y[i]
        for j in range(cols):
            m_slope[j] += (-2/M)*(Y[i] - (m*x).sum())*x[j]
            
    new_m = m - (learning_rate * m_slope)
    return new_m
```

```python
def cost(points, m):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i]
        y = Y[i]
        total_cost += (1/M)*((y-((m*x).sum()))**2)
        
    return total_cost
```

```python
def gd(points, learning_rate, num_iterations):
    cols = points.shape[1]
    m = np.zeros(cols)
    
    for i in range(num_iterations):
        m = step_gradient(points, learning_rate, m)
        print(i, "Cost: ", cost(points, m))
        
    return m
```

```python
def run():
    learning_rate = 0.156
    num_iterations = 100
    m = gd(data, learning_rate, num_iterations)
    
    for i in range(len(m)):
        print(m[i], " ")
        
    return m
        
m = run()
```

# Now on test data predict the y (Method 1)

```python
# Load the data using pandas and insert a column for the constant
test_df = pd.read_csv("test.csv", header = None)
test_df['c'] = np.ones(test_df.shape[0])
test_df
```

```python
# Convert the test_df into np array and find y_test_predicted
test_points = np.array(test_df)
y_test_predicted = []
M = test_points.shape[0]
for i in range(M):
    x = test_points[i]
    output = (m*x).sum()
    y_test_predicted.append(output)
    
y_test_predicted = np.array(y_test_predicted)
print(y_test_predicted.shape)
print(y_test_predicted)
```

```python
# Save the y_test_predicted in the csv file
np.savetxt('y_test_predicted.csv', y_test_predicted, delimiter=',',fmt='%1.5f')
```
